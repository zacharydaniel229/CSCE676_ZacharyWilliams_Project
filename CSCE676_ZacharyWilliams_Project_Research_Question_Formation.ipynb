## A. Research Questions

### Research Question 1
Do similar reviews correlate with similar ratings?
- Data mining task type: Text mining
- Relevant algorithms: word2vec (course), BERT (external)
- Evaluation criteria: cosine similarity

### Research Question 2
Can we predict the ranking of items within a category based on words used in the reviews?
- Data mining task type: Text mining, large-scale ml
- Relevant algorithms: word2vec (course), BERT (external), randomForests (course), Logistic regression (external)
- Evaluation criteria: accuracy, f1 score, MSE, R^2 score

### Research Question 3
What clusters emerge from review text embeddings?
- Data mining task type: Text mining, clustering
- Relevant algorithms: word2vec (course), BERT (external), DBSCAN (course), k-means (course)
- Evaluation criteria: silhouette score
